```python
# Question 1
series1 = pd.Series([4, 65, 436, 3, 9])
series2 = pd.Series([7, 0, 3, 897, 9])

#Compare the two series
comparison = series1 == series2

# Print out the comparison result
print(comparison)
```

```python
# Question 2
one = pd.Series([4, 65, 436, 3, 9])
two = pd.Series([7, 0, 3, 897, 9])

# Addition
add = one + two

# Subtraction
subtract = one - two

# Multiplication
multiply = one * two

# Division 
divide = one / two

## Print out the result
print("Addition:")
print(add)
print("Subtraction:")
print(subtract)
print("Multiplication: ")
print(multiply)
print("Division: ")
print(divide)
```

```python
# Question 5
df = pd.read_csv('csv/AfricaCupofNationsMatches.csv')
# Extract the 'HomeTeamGoal' as a series
home_goals = df['HomeTeamGoals']

# Find the most frequent value 
most_frequent = home_goals.value_counts().idxmax()

# Replace all the other values with 'Other'
home_goals = home_goals.apply(lambda x: x if x ==most_frequent else 'Other')

# Display the update series

print(home_goals)
```
